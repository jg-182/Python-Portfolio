```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://pt.wikipedia.org/wiki/Lista_das_maiores_empresas_de_Portugal'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')
```

```python
table = soup.find('table')
print(table)
```

```python
pt_headers = table.find_all('th')
print(pt_headers)
```

```python
pt_headers_names = [ header.text.strip() for header in pt_headers ]
print(pt_headers_names)
```

```python
pt_headers_names.remove("Sede")
print(pt_headers_names)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = pt_headers_names)
df
```

```python
columns_data = table.find_all('tr')
print(columns_data)
```

```python
for row in columns_data:
    rows_data = row.find_all('td')
    pt_data = [ data.text.strip() for data in rows_data ]
    print(pt_data)
```

```python
for row in columns_data[2:]:
    rows_data = row.find_all('td')
    pt_data = [ data.text.strip() for data in rows_data ]
    
    length = len(df)
    df.loc[length] = pt_data
```
